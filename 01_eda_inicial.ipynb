```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from scipy.stats import zscore

pd.set_option("display.max_columns", None)

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/processed/properties_clean.csv")

df.head()
```

```python
print("Número de registros:", df.shape[0])

print("Número de variables:", df.shape[1])

print("\nTipos de datos:")

print(df.dtypes)
```

```python
nulls = df.isnull().sum()

null_percentage = (

    df.isnull().sum() / len(df) * 100

).round(2)

nulls_df = pd.DataFrame({

    "null_count": nulls,

    "null_percentage": null_percentage

})

nulls_df
```

```python
text_columns = df.select_dtypes(include=["object"]).columns

blank_counts = {}

for col in text_columns:
    blank_counts[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .eq("")
        .sum()
    )

blank_df = pd.DataFrame.from_dict(
    blank_counts,
    orient="index",
    columns=["blank_values"]
)

blank_df
```

```python
print("Duplicados exactos:")
print(df.duplicated().sum())

print("\nDuplicados por source + property_id:")
print(
    df.duplicated(
        subset=["source", "property_id"]
    ).sum()
)
```

```python
print("Duplicados exactos:")
print(df.duplicated().sum())

print("\nDuplicados por source + property_id:")
print(
    df.duplicated(
        subset=["source", "property_id"]
    ).sum()
)
```

```python
df.describe()
```

```python
categorical_columns = [
    "source",
    "city",
    "country",
    "neighborhood",
    "property_type"
]

for col in categorical_columns:
    print(f"\n{col}")
    print(df[col].nunique())
```

```python
numeric_columns = [
    "price_eur",
    "area_m2",
    "price_per_m2"
]

for col in numeric_columns:

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    print(f"\n{col}")
    print(f"Outliers detectados: {len(outliers)}")
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    df.isnull(),
    cbar=False,
    yticklabels=False
)

plt.title("Mapa de calor de valores nulos")

plt.show()
```

```python
numeric_columns = [
    "price_eur",
    "area_m2",
    "price_per_m2"
]

for col in numeric_columns:

    plt.figure(figsize=(8,5))

    sns.histplot(
        df[col],
        kde=True
    )

    plt.title(f"Distribución de {col}")

    plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x="city",
    y="price_per_m2",
    data=df
)

plt.title("Precio por m² por ciudad")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df["neighborhood"].value_counts().plot(
    kind="bar"
)

plt.title("Frecuencia de barrios")

plt.xlabel("Barrio")

plt.ylabel("Número de propiedades")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x="area_m2",
    y="price_eur",
    hue="city",
    data=df
)

plt.title("Relación entre superficie y precio")

plt.show()
```
